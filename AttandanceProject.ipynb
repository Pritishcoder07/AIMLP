```python
attendance=pd.read_csv('attendance_100_real_names.csv')
```

```python
attendance
```

```python
attendance['Attendance_percentage']=attendance['Attended_Classes']/attendance['Total_Classes']*100
```

```python
attendance
```

```python
Less_Attendance=attendance[attendance['Attendance_percentage']<75]
```

```python
Less_Attendance
```

```python
labels = ["Above 75%", "Below 75%"]
sizes = [
    len(attendance[attendance['Attendance_percentage'] >= 75]),
    len(attendance[attendance['Attendance_percentage'] < 75])]


plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Attendance Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.bar(labels, sizes, color=['green', 'red'])
plt.xlabel('Attendance Category')
plt.ylabel('Number of Students')
plt.title('Attendance Distribution (Above/Below 75%)')
plt.show()
```
