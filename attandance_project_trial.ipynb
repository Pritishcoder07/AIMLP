```python
df = pd.read_csv("attendance_100_students.csv")
df
```

```python
df["Attendance_Percentage"] = (
    df["Attended_Classes"] / df["Total_Classes"]
) * 100

df
```

```python
low_attendance = df[df["Attendance_Percentage"] < 75]
low_attendance
```

```python
df["Attendance_Percentage"].describe()
```

```python
plt.figure()
plt.bar(df["Name"], df["Attendance_Percentage"])
plt.xlabel("Student Name")
plt.ylabel("Attendance Percentage")
plt.title("Student Attendance Analysis")
annot=True

```

```python
labels = ["Above 75%", "Below 75%"]
sizes = [
    len(df[df["Attendance_Percentage"] >= 75]),
    len(df[df["Attendance_Percentage"] < 75])
]

plt.figure()
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Attendance Distribution")
plt.show()
```
